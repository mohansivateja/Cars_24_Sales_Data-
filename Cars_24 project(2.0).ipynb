```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
Brand = []
year = []
Driven_by_km = []
Fuel_type = []
Transmission = []
Variant = []
EMI_per_month = []
Price = []

for i in l:
    print(f"https://www.cars24.com/buy-used-cars-{i}/")
    
    connection = requests.get(f"https://www.cars24.com/buy-used-cars-{i}/")
    text = connection.text
    html_readable_format = BeautifulSoup(text, 'html.parser')
    html_readable_format

#Brand
    for item in range(len(html_readable_format.find_all("h3",class_="_11dVb"))):
         extracted_text=html_readable_format.find_all("h3",class_="_11dVb")[item].text
         Brand_extracted = re.findall(pattern=r"^\d+\s(\w+)",string=extracted_text)
         if len(Brand_extracted)>0:
              Brand.append(Brand_extracted)
    Brand

#year
    for item in range(len(html_readable_format.find_all("h3",class_="_11dVb"))):
        extracted_text=html_readable_format.find_all("h3",class_="_11dVb")[item].text
        year_extracted = re.findall(pattern=r"\d{4}",string=extracted_text)
        if len(year_extracted)>0:
            year.append(year_extracted)
    year

#Driven_by_kms
    for item in range(len(html_readable_format.find_all("ul",class_="_3J2G-"))):
        extracted_text=html_readable_format.find_all("ul",class_="_3J2G-")[item].text
        Driven_by_km_extracted = re.findall(pattern=r"\d+,\d+,\d+ km|\d+,\d+ km",string=extracted_text)
        if len(Driven_by_km_extracted)!=0:
            Driven_by_km.append(Driven_by_km_extracted)
        else:
            Driven_by_km.append(np.nan)
    Driven_by_km

#Fuel_type
    for item in range(len(html_readable_format.find_all("ul",class_="_3J2G-"))):
        extracted_text=html_readable_format.find_all("ul",class_="_3J2G-")[item].text
        Fuel_type_extracted = re.findall(pattern=r"Petrol|Diesel",string=extracted_text)
        if len(Fuel_type_extracted)>0:
            Fuel_type.append(Fuel_type_extracted)
        else:
            Fuel_type.append(np.nan)
    Fuel_type

#Transmittion
    for item in range(len(html_readable_format.find_all("ul",class_="_3J2G-"))):
        extracted_text=html_readable_format.find_all("ul",class_="_3J2G-")[item].text
        Transmission_extracted = re.findall(pattern=r"Manual|Automatic",string=extracted_text)
        if len(Fuel_type_extracted)>0:
            Transmission.append(Transmission_extracted)
        else:
            Transmission.append("Nan")

    

#Variant
    for item in range(len(html_readable_format.find_all("h3",class_="_11dVb"))):
        extracted_text=html_readable_format.find_all("h3",class_="_11dVb")[item].text
        Variant_extracted = re.findall(pattern=r"^\d+\s\w+\s(\w.*)",string=extracted_text)
        if len(Variant_extracted)>0:
            Variant.append(Variant_extracted)

    Variant


#EMI_Starts
    for item in range(len(html_readable_format.find_all("div",class_="-EPGU"))):
        extracted_text=html_readable_format.find_all("div",class_="-EPGU")[item].text
        EMI_per_month_extracted = re.findall(pattern=r"₹\d+,\d+",string=extracted_text)
        if len(EMI_per_month)<401:
            EMI_per_month.append(EMI_per_month_extracted)
       
    EMI_per_month

#Pricing
    for item in range (len(html_readable_format.find_all("div",class_="_2KyOK"))):
        Price.append(html_readable_format.find_all("div",class_="_2KyOK")[item].text.split()[0])
    Price

```

```python
cars_24_data = pd.DataFrame(
    {"Brand": Brand,
     "year":year,
     "Driven_by_km":Driven_by_km, 
     "Fuel_type":Fuel_type, 
     "Transmission":Transmission,
     "Variant":Variant,
     "EMI_per_month(₹)":EMI_per_month, 
     "Price(₹)":Price   }
)

cars_24_data['Brand']=[i[0] for i in cars_24_data['Brand']]
cars_24_data['year']=[i[0] for i in cars_24_data['year']]
#cars_24_data['Driven_by_km']=[i[0] for i in cars_24_data['Driven_by_km']]
#cars_24_data['Fuel_type']=[i[0] for i in cars_24_data['Fuel_type']]
cars_24_data['Transmission']=[i[0] for i in cars_24_data['Transmission']]
cars_24_data['Variant']=[i[0] for i in cars_24_data['Variant']]
#cars_24_data['EMI_per_month']=[i[0] for i in cars_24_data['EMI_per_month']]
```

```python
cars_24_data["Price(₹)"] = cars_24_data["Price(₹)"].str.replace("₹","")
```

```python
columns_to_clean = ['Driven_by_km', 'Fuel_type', 'EMI_per_month(₹)']

for column in columns_to_clean:
    cars_24_data[column] = cars_24_data[column].astype('str').str.replace('₹','').str.replace('km','').str.replace('[', '').str.replace(']','' ).str.replace("'","") .str.replace(',','')
```
